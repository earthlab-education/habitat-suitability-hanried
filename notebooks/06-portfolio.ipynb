## **Introduction:**

In this project, I modeled future habitat suitability of the Siskiyou and Los Padres National Forests for the coast redwood tree (*Sequoia sempervirens*). The coast redwood currently lives in both national forests as of April 2025. I wanted to focus on the coast redwood because the tree's current habitat in the continental United States is already narrow and specific, so if the current habitat becomes unsuitable, that could drastically affect the trees as they don't have many other places they currently occupy. Also, I think the coast redwoods are magnificent trees and an iconic part of the west coast, so I want them to continue to be able to survive in their habitat.

I decided to compare two future time periods: 2036-2065 and 2066-2095. The first time period allowed for a summary of climate in 2050 and the second time period allowed for a summary of climate in 2080. As it is currently 2025, I was curious about what habitat suitability could look like in 25 years and 55 years.

To judge the future suitability of the Siskiyou and Los Padres National Forests, I downloaded and wrangled soil pH, elevation, and predicted precipitation data. I then harmonized those data and created fuzzy logic models. The fuzzy logic models were multiplied together via raster multiplication. I applied a habitat suitability threshold of 0.5 to the raster products to determine habitat suitability for each scenario.

I decided to use a fuzzy logic model because it allows for nuances in the suitability thresholds of each variable - soil pH, elevation, and predicted precipitation. For example, the coast redwood can live in elevation from sea level to 3,000 feet, however it prefers elevations between 100 and 2,500 feet (Roy, 1966). With a fuzzy logic model, I was able to take the optimal range and tolerable range into account. If I had solely used a model based in boolean logic, I would just have used one of those ranges, either sea level to 3,000 feet or 100 to 2,500 feet. If the elevation was outside of whichever range I picked, the model would have said "There's no way a coast redwood could live here" when maybe the tree can, it just isn't preferable.

In the end, I had 16 habitat suitability models. There were 8 models for the Siskiyou National Forest and 8 models for the Los Padres National Forest. Within the 8 models for each site, there were four models for each time period. Within those 4 models for each site/time period combination, there was one model for each climate scenario.

## **Species Description:**

According to the USDA PLANTS Database profile of the coast redwood, the tree is a perennial so it is actively growing in the spring and summer. The tree is also fire resistant. In terms of soil, the tree tolerates a pH range of 5.0 - 7.0, is adapted to coarse and medium textured soils, and has a minimum root depth of 40 inches. Redwoods require a minimum of 180 frost free days and a precipitation range of 20 - 200 inches (508 mm - 5080 mm).

According to Roy's 1966 paper, the optimal soil pH is 6.5 and the tolerable range is 5.0 - 7.5. Additionally, as mentioned above, coast redwoods can grow at elevations from sea level to 3,000 feet, although generally they live between 100 and 2,500 feet.

The Fire Effects Information System (FEIS) website offers further information. The coast redwood is found on the coast of California and southwestern Oregon. The Northern boundary is the Chetco River in the Siskiyou Mountains near the Oregon/California border. The southern boundary of the redwood's habitat is Salmon Creek Canyon in Santa Lucia Mountains in Monterey County, California. The coast redwood exists in sites with cool and rainy winters and dry summers. The average annual precipitation of preferred redwood sites is 70 inches, 90% of which occurs between October and May. Additionally, fog can play an extremely important role in the redwood's survival. Heavy fog in the summers offsets lack of precipitation, reduces evapotranspiration and adds soil moisture. If a redwood lives somewhere with less fog, more precipitation must occur. The FEIS agrees with Roy's 1966 paper that redwoods tolerate elevation from sea level to 3,000, however FEIS lists a preferred elevation of 100 to 2,320 feet. FEIS describes that many beings use the redwood forests for habitats, such as the spotted owl, marbled murrelet, Roosevelt elk, and black-tailed deer. The Tanoak (Lithocarpus densiflorus) and Pacific madrone (Arbutus menziesii) are competitors to the redwood and there are some insects and rodents that can damage the trees. Finally, redwood trees can grow to be over 300 feet tall. (Griffith, 1992)

## **Site Descriptions:**

I decided to focus on the Siskiyou and Los Padres National Forests because they are at the northern and southern ends of the coast redwood's current habitat, respectively.

The Siskiyou National Forest was combined with the Rogue River National Forest in 2004 (Rogue River-Siskiyou National Forest - Home, n.d.), however I just focused on the Siskiyou National Forest because Rogue River and Siskiyou are separate sites in the US Forest Service National Forests dataset I used to get the forest boundaries.

#### Rogue River-Siskiyou National Forest details (Rogue River-Siskiyou National Forest - Home, n.d.):

* Located on the southwestern border of Oregon with small parts in California
* 1.8 million acres
* Includes the Siskiyou Mountains - elevation from sea level to 7,533 feet
* On the coast, there is the Coast Range temperate rain forest, 60-100 inches of rain, mainly falling from October to June. Moving inland, annual precip is about 30 inches lower in the Siskiyou Mountians, where most of the precipitation is during October-April, both as rain and snow in higher elevations. In the summer, coastal temperatures usually have a max of 75 degrees celsius. Temperatures inland in the summer are more in the 80s and 90s.

#### Los Padres National Forest (Los Padres National Forest - About the Area, n.d.):

* 1.75 million acres, 220 miles long north to south 
* Serves San Francisco Bay Area, LA Metro area, southern San Joaquin Valley, among other communitites; "supplies a substantial portion of the water needs of several downstream communities"
* Varied ecosystems: semi-desert inland, redwood forest on coast
* About 468 species of fish and wildlife; 23 threatened and endangered animals
* This is an important area for wildlife due to urban development around the National Forest
* Part of the California condor Recovery Program
* Since 1912, 2.3 million acres in the forest were burned by wildfire, 25,000 acres per year; average annual wildfire events have increased over the last 60 years, due to "urban encroachment, expanded recreational use of the forest, and old-age chaparral"; chaparral makes up a majority of the land burned
* Some of the land is used for livestock grazing
* Locatabe minerals: gold, copper, zinc, chrome, antimony, silver, uranium, mercury, gypsum, limestone and bentonite
* Leasable minerals: geothermal energy, phosphates, oil and gas. Oil and gas are produced by the forest in commercial amounts

#### Soil Data: POLARIS

I downloaded soil pH data from the POLARIS dataset. The resolution of this data is 1 arcsec (about 30 meters). I downloaded mean soil values for a soil depth of 100-200 cm since the minimum root depth of the redwood is 40 inches (about 102 cm) (USDA Plants Database Plant Profile Characteristics, n.d.).

#### Elevation Data: SRTM

Elevation data was downloaded from the Shuttle Radar Topography Mission (SRTM) Dataset using the earthaccess API. Specifically, I used the SRTMGL1 NASA Shuttle Radar Topography Mission Global 1 arc second V003 dataset. The resolution of this data is about 30 meters and contains data tiles of elevation measurements. The SRTM datasets are produced by the National Aeronautics and Space Administration (NASA), National Geospatial-Intelligence Agency (NGA), and the German and Italian space agencies. (NASA JPL, 2013)

## **Method:**

This project had 8 major steps:

1. Use GBIF and national forest data to pick the species and sites of interest.
2. Research the species, sites, and data sources.
3. Download and wrangle the POLARIS soil data for each site.
4. Download and wrangle the SRTM elevation data for each site.
5. Download and wrangle the precipitation data for the four different climate models from the MACA v2 Climate Datasets.
6. Harmonize the soil, elevation, and precipitation data.
7. Use SciKit-Fuzzy to create fuzzy logic models out of trapezoidal membership functions. Use raster multiplication to multiply the fuzzy logic models together. Apply a habitat suitability threshold to the raster products to determine habitat suitability.
8. Create plots to visualize habitat suitability.
